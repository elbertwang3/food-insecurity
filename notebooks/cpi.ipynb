```python
cpi.get(1913)
```

```python
cpi_data = []
cpi_data = pd.DataFrame(cpi_data)
cpi_data = pd.DataFrame(columns=('year','housing','housing_inf','food','food_inf', 'housing_national', 'housing_national_inf', 'national', 'national_inf'))
years = list(range(1982, 2018))
print (years)
#years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
cpi_data.year = years

cpi_data
```

```python
housing = []
for y in years:
    housing.append(cpi.get(y, items="Housing", area="San Francisco-Oakland-Hayward, CA"))
cpi_data.housing = housing
```

```python
housing_inf = []
for year in years:
    housing_inf.append(cpi.inflate(cpi.get(year, items="Housing", area="San Francisco-Oakland-Hayward, CA"),year))
cpi_data.housing_inf = housing_inf
```

```python
food_inf = []
for year in years:
    food_inf.append(cpi.inflate(cpi.get(year, items="Food", area="San Francisco-Oakland-Hayward, CA"), year))
cpi_data.food_inf = food_inf
```

```python
food = []
for year in years:
    food.append(cpi.get(year, items="Food", area="San Francisco-Oakland-Hayward, CA"))
cpi_data.food = food
```

```python
housing_national = []
for y in years:
    print(cpi.get(y, items="Housing"))
    housing_national.append(cpi.get(y, items="Housing"))
cpi_data.housing_national = housing_national
```

```python
housing_national_inf = []
for y in years:
    print(cpi.inflate(cpi.get(y, items="Housing"),year))
    housing_national_inf.append(cpi.inflate(cpi.get(y, items="Housing"),year))
cpi_data.housing_national_inf = housing_national_inf
```

```python
national = []
for y in years:
    print(cpi.get(y))
    national.append(cpi.get(y))
cpi_data.national = national
```

```python
national_inf = []
for y in years:
    national_inf.append(cpi.inflate(cpi.get(y),year))
cpi_data.national_inf = national_inf
```
